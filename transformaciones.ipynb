*FUNCIONES

```python
# Esta función convierte a minuesculas los df de tipo objeto

def df_Tolower(df_o):

    lowers = df_o.select_dtypes('O')

    objcolumns = list(lowers.columns)

    for columnaName in objcolumns:
        lowers[columnaName] = lowers[columnaName].str.lower()
        df_o[columnaName] = lowers[columnaName]
```

```python
#La finalidad de esta función es corregir algunos valores de la columna "duration" que se encuentran en "rating"

def huluCorreciones (x):
    
    if('seasons' in x  or 'season' in x or 'min' in x):
        return x
    else:
        return np.nan
```

*HULU

```python
data_hulu=pd.read_csv("./datasets/hulu_titles-score.csv", parse_dates=['date_added'])
data_hulu.info()
```

```python
list_showIdHulu = data_hulu['show_id'].tolist() 

for valor in range(0,len(list_showIdHulu)):
    list_showIdHulu[valor] = 'h'+list_showIdHulu[valor]
    
valores_IdHulu = tuple(list_showIdHulu)
data_hulu.insert(0,"id", value = valores_IdHulu)
data_hulu.head()
```

```python
data_hulu= data_hulu.fillna({'rating':'g'})
```

```python
df_Tolower(data_hulu)
data_hulu.head()
```

```python
data_hulu.loc[data_hulu['duration'].isnull() == True] #479 rows de la columna 'duration' son NaN
```

```python
#Creamos la columna 'correcciones' esta tendrá los resultados de aplicar la función 'huluCorreciones
data_hulu['correciones']=data_hulu.rating.apply(lambda x: huluCorreciones(x))

data_hulu.correciones.info() # 292 de los 479 NaN que hay en 'duration' en realidad se encuentran en la columna 'rating'
```

```python
#Creamos dos objetos tipo series que contendran la columna 'duration' y 'correciones' para aplicarle el metodo combine
sh_duration = pd.Series(data_hulu.duration)
sh_correciones = pd.Series(data_hulu.correciones)
data_hulu.duration = sh_duration.combine_first(sh_correciones) #agregamos a la columna 'duration' el resultado de combinar

data_hulu.drop(columns=['correciones'], inplace= True) #dropeamos la columna 'correciones'
data_hulu.info()
```

```python
# Corregimos los 292 valores de la columna 'rating':
# como quedan en NaN, atendiendo a los requerimientos para esta columna los reemplazamos por g

#data_hulu[data_hulu.rating.str.contains('min') | data_hulu.rating.str.contains('season')]

data_hulu.loc[data_hulu.rating.str.contains('min'),'rating'] = 'g'
data_hulu.loc[data_hulu.rating.str.contains('season'),'rating'] = 'g'
data_hulu.loc[data_hulu.rating.str.contains('seasons'),'rating'] = 'g'

# data_hulu[data_hulu.rating.str.contains('season')] -> Puede verificar la info con esa linea cambiado el parametro 'season' por el que necesite

# mask=(data_hulu['id'] == 'hs2960') | 
#                                    |  -> ejemplo de verificar que todo se encuentra bien
# data_hulu.loc[mask, :]             |
```

```python
data_hulu_duration = data_hulu.duration.str.split(expand=True)

data_hulu_duration.rename(columns = {0:'duration_int',
                                    1:'duration_type'}, inplace = True)

data_hulu.insert(10,column='duration_int',value = data_hulu_duration.duration_int)
data_hulu.insert(11,column='duration_type',value = data_hulu_duration.duration_type)

data_hulu['duration_int'] = data_hulu['duration_int'].astype('Int64')

data_hulu['duration_type'].replace(['seasons'],'season', inplace=True)

# mask=(data_ama['duration_type'] == 'seasons') |
#                                               |->pruebe que esta normalizado
# data_net.loc[mask, :]                         |

data_hulu.head()
```
